```python
# ELU(x) = alpha(e^x - 1) if x < 0
#        = x              if x > 0
# 
import numpy as np 
import matplotlib.pyplot as plt 

def elu(x, alpha=0.1):
    return np.where(x <= 0, alpha*(np.exp(x) - 1), x)

x = np.arange(-10, 10, 0.5)
y = elu(x)

plt.plot(x, y)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# swish(x) = x/(1 + e ^ (-beta * x))
#          = x * sigmoid(belta*x)
import numpy as np
import matplotlib.pyplot as plt

def swish(x, beta = 1):
    return x/(1 + np.exp(-beta * x))

def swish_derivative(x, beta = 1):
    sigmoid = 1/(1 + np.exp(- beta * x))
    return sigmoid * (1 + beta * x * (1 - sigmoid))

x = np.arange(-10, 10, 0.5)
y = swish(x)
y1 = swish_derivative(x)

plt.plot(x, y, label="swish", color='blue', linewidth=2)
plt.plot(x, y1, label="swish_derivation", color='red', linewidth=2)
plt.grid(True, linestyle='--', alpha = 0.5)
plt.legend(fontsize=12)
plt.show()
```

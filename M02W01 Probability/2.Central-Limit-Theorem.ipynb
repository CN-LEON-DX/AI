```python
import seaborn as sns
import cv2
import random

img = cv2.imread("img.png", 0)
height, width = img.shape
print(height, width)
```

```python
num_of_sample = 10000
sz = 500
means = []

for _ in range(num_of_sample):
    mean = 0
    for _ in range(sz):
        x = random.randint(0, height-1)
        y = random.randint(0, width-1)
        mean += img[x][y]
    mean = mean/sz
    means.append(mean)
```

```python
sns.displot(means, kde=True)
```

```python
import random
import math

num_of_samples = 10000
sample_size = 30
side = 500

means = []
for _ in range(num_of_samples):
    mean = 0
    for _ in range(sample_size):
        x = random.randint(0, side-1)
        y = random.randint(0, side-1)
        distance = math.sqrt(x*x + y*y)
        mean = mean + distance
    mean = mean/sample_size
    means.append(mean)
```

```python
sns.displot(means, kde=True)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("img.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)

sh = img.shape
print(sh)


```

```python
# blur: 100 - 150 h, 400 -> 500 w

blur = img.copy()

y1, y2 = 100, 270
x1, x2 = 370, 550

# crop
cr = blur[y1:y2, x1:x2]

plt.imshow(cr)
```

```python
r, g, b = 0, 0, 0
row, col = cr.shape[:2]

for i in range(row):
    for j in range(col):
        pixel = cr[i, j]
        r += int(pixel[0])
        g += int(pixel[1])
        b += int(pixel[2])

print(r, g, b)
sz = row * col
r, g, b = r // sz, g // sz, b // sz

avg_color = [r, g, b]
print(avg_color)
# then assign backto img

blur[y1:y2, x1:x2] = avg_color

plt.imshow(blur)
plt.show()
```

```python
kernel = np.ones((6, 6), np.float32) / 25.0

dst = cv2.filter2D(cr, -1, kernel)

plt.subplot(121), plt.imshow(cr), plt.title('Original')
plt.subplot(122), plt.imshow(dst), plt.title('Blurred')
plt.show()
```

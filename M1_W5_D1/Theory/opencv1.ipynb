```python
import cv2
from google.colab.patches import cv2_imshow

def read_and_display_grayscale(image_path):
    # read the image in grayscale mode
    grayscale_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    resized_image = cv2.resize(grayscale_image, (500, 500), interpolation=cv2.INTER_AREA)

    cv2_imshow(resized_image)


read_and_display_grayscale('mango.png')
```

```python
def read_and_display_rgb(image_path):
  color_image = cv2.imread(image_path, cv2.IMREAD_COLOR)
  resize = cv2.resize(color_image, (500, 500), interpolation=cv2.INTER_AREA)
  cv2_imshow(resize)

read_and_display_rgb('mango.png')
```

```python
# increase the brightness of a grayscale image

def increase_brighness(image_path, beta_value):
  # load image
  img = cv2.imread(image_path, cv2.IMREAD_COLOR)
  # increase by adding beta value to all pixel intensities
  # output = alpha * input + beta
  brightness_image = cv2.convertScaleAbs(img, alpha=1, beta=beta_value)

  resize = cv2.resize(brightness_image, (500, 500), interpolation=cv2.INTER_AREA)
  cv2_imshow(resize)

increase_brighness('mango.png', 50)

```

```python
# draw a bounding box

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('mango.png', 1)
img = cv2.resize(img, (500, 500), interpolation=cv2.INTER_AREA)

height, width, channel = img.shape

x0, y0 = 120, 200
x1, y1 = 275, 400

# BGR for opencv when read image
color = np.array([0, 0, 255])

# draw 4 edge of object
# draw width
img[y0, x0:x1, :] = color
img[y1, x0:x1, :] = color
# draw height
img[y0:y1, x0, :] = color
img[y0:y1, x1, :] = color

cv2.imwrite('output_mango.png', img)

res = cv2.imread('output_mango.png', 1)
cv2_imshow(res)
```

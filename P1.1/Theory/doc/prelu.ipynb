```python
# PReLU(x) = alpha*x if x <= 0
# PReLU(x) = x       if x >  0

import numpy as np
import matplotlib.pyplot as plt 

def PReLU(x, alpha = 0.1):
    return np.where(x <= 0, alpha * x, x)

x = np.arange(-10, 10, 0.5)
y = PReLU(x)

plt.plot(x, y)
plt.grid(True, linestyle='--', alpha = 0.5)
plt.show()
```

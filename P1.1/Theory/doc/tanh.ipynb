```python
# tanh function 
# tanh(x) = 2/(1 + e^(-2x)) - 1
# (-1, 1)

import numpy as np
import matplotlib.pyplot as plt

plt.axhline(0, color='black', linewidth=0.5)  
plt.axvline(0, color='black', linewidth=0.5)

def tanh(x):
    return 2/(1 + np.exp(-2*x)) - 1
x = np.arange(-10, 10, 0.5)
y = tanh(x)

plt.plot(x, y)
plt.xlabel("X")
plt.xlabel("Y")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# sigmoid(x) = 1 / (1 + e^(-x))
# (0, 1)

import math
import matplotlib.pyplot as plt

def sigmoid(x) :
    return 1/(1 + math.exp(-x))
x = []
y = []
# can't ussing step 0.5 here for float number
# # just using numpy for this 
for i in range(-10, 10, 1):
    x.append(i)
    y.append(sigmoid(i))

plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("y")

plt.show()
```

```python
# using numpy for process array 

import numpy as np 
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1 + np.exp(-x))
x = np.arange(-10, 10, 0.5)
print(x)

y = sigmoid(x)

plt.plot(x, y)
plt.xlabel("X")
plt.xlabel("Y")
plt.show()
```
